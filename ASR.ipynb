```python
import os, librosa

```

```python
# Set the root directory
root_dir = "WAV"
# Iterate over all subdirectories in the root directory
for subdir, dirs, files in os.walk(root_dir):
    # Iterate over all files in the current subdirectory
    for dirs in dirs:
        subfolder = os.path.join(root_dir,dirs)
        for file in files: 
            file_path = os.path.join(subfolder, file)
            signal, sr = librosa.load(file_path)
            # Extract MFCCs and transpose the matrix so that the features 
            # are the columns
            mfccs = librosa.feature.mfcc(signal, sr, n_mfcc=13).T
```

```python
# Set the root directory
root_dir = "WAV"
for folder in os.listdir(root_dir):
    subfolder = os.path.join(root_dir, folder)
    # Iterate over all files in the current subdirectory
    for file in os.listdir(subfolder):
        file_path = os.path.join(subfolder, file)
        signal, sr = librosa.load(file_path)
        # Extract MFCCs and transpose the matrix so that the features 
        # are the columns
        mfccs = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=13).T
```
